```python
import numpy as np
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import seaborn as sns
```

```python
def read_textfiles(path):
    poss_columns = ['Lenkradwin', 'Lenkmoment', 'Fahrgeschw', 'Querbeschl', 'Giergeschw', 'Wankwinkel', 'Radius']
    files = sorted(os.listdir(path))
    print("No. of files to be processed are: ", len(files))
    count  = 0
    data = []
    dup_path = path
    for each_file in files:
        dup_path += each_file
#         print(dup_path)
        print(count, end=',')
        data.append(pd.read_csv(dup_path, sep = '\t', engine = 'python', decimal = ',', skiprows = 40, usecols = poss_columns)[2: ].apply(pd.to_numeric))
        count += 1
        dup_path = path
    print("No. of files processed are: ", count)
    return data
```

```python
file_data = read_textfiles('/home/loki/MAS/R_D/Moses/Moses_Data/')
```

```python
# kendall = file_data[0].corr(method = 'kendall')
# pearson = file_data[0].corr(method = 'pearson')
spearman = file_data[0].corr(method = 'spearman')
plt.figure(figsize=(20,20))
sns_map = sns.heatmap(spearman, annot = True)
# plt.savefig("fig.png")
plt.show()
```

## Autoencoder implementation:

```python
from tensorflow.keras.layers import Input, Dense, LSTM, TimeDistributed, RepeatVector
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import regularizers, callbacks, optimizers
```

```python
input_value = Input(shape = (7, ))
encoded_value_1 =  Dense(4, activation='relu', activity_regularizer=regularizers.l2(0.01))(input_value)
encoded_value = Dense(3, activation='relu', activity_regularizer=regularizers.l2(0.01))(encoded_value_1)
decoded_value_1 = Dense(4, activation='relu', activity_regularizer=regularizers.l2(0.01))(encoded_value)
decoded_value = Dense(7, activation='sigmoid', activity_regularizer=regularizers.l2(0.01))(decoded_value_1)


autoencoder = Model(input_value, decoded_value)
# Seperate encoder for latent space representation
encoder = Model(input_value, encoded_value)

sgd_opt =  optimizers.SGD(lr=0.1, momentum=0.1)

early_stopping = callbacks.EarlyStopping(monitor='val_accuracy', patience=5, mode='max')
```

```python
autoencoder.compile(optimizer=sgd_opt, loss='mean_squared_error', metrics = ['accuracy'])
autoencoder_history = autoencoder.fit(new_pd, new_pd, 
                epochs = 10, batch_size = 250, 
               shuffle = False, 
                validation_data = (valid_pd, valid_pd),
                callbacks=[early_stopping])
```

## LSTM Autoencoder implementation:

```python
model = Sequential()
model.add(LSTM(7, activation='relu', input_shape= (len(new_pd), 7 ), return_sequences=True))
model.add(LSTM(4, activation='relu', return_sequences=True))
model.add(LSTM(3, activation='relu', return_sequences=True))
model.add(LSTM(4, activation='relu', return_sequences=True))
model.add(LSTM(7, activation='sigmoid', return_sequences=True))
model.add(TimeDistributed(Dense(7)))
model.compile(optimizer = 'adam', loss = 'mse')
model.summary()
```
