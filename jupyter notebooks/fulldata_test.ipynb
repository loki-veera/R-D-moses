### Read the data

```python
def read_files(file_names):
    data = []
    for each_file in file_names:
        path = '/home/loki/MAS/R_D/Moses/Data/Auto1_F1_SK/'
        path = path + each_file
        df = pd.read_csv(path, sep="\t", engine= 'python', decimal = ',')
        data.append(df)
    return data
```

```python
data = read_files(files)
new_data = []
new_data.append(data[0])
new_data.append(data[2])
new_data.append(data[4])
```

### Plot the features

```python
print(list(data[0].columns))
# poss_features = ['Radius', 'Fahrgeschw', 'Lenkradwin', 
#                  'Querbeschl', 'Giergeschw', 
#                  'Wankwinkel', 'Lenkmoment', 'Gierwinkel']
poss_features = ['Lenkradwin',  'Lenkmoment', 'Wankwinkel',
                 'Radius', 'Gierwinkel',
                 'Giergeschw', 'Querbeschl', 'Fahrgeschw']
```

```python
def get_feature_values(data, feature):
    param_values = data[feature][2:]
#     param_values = [float(each_val.replace(',', '.')) for each_val in param_values]
    return param_values
def plot_param(feature, units):
    plt.figure(figsize=(15, 10))
    count = 0
    y_label = feature+" ("+ units +")"
    for each_file in new_data:
        count+=1
        param_values = get_feature_values(each_file, feature)
        label_t = "Test run - "+str(count)
        plt.plot(param_values, label=label_t)
        plt.gca().legend()
        plt.xlabel("Sample time(sec)", fontsize=20)
        plt.ylabel(y_label, fontsize=20)
        title_val = feature+" ("+ units +")" + " Vs Sample time(sec)" 
        plt.title(title_val, loc="right", fontsize=20)
    plt.show()
```
