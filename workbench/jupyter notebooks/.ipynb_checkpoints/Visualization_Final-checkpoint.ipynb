```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Read the text files from the path argument
def read_textfiles(path):
    poss_columns = ['Lenkradwin', 'Lenkmoment', 'Fahrgeschw', 'Querbeschl', 'Giergeschw', 'Wankwinkel', 'Radius']
    files = sorted(os.listdir(path))
    print("No. of files to be processed are: ", len(files))
    count  = 0
    data = []
    dup_path = path
    for each_file in files:
        dup_path += each_file
#         print(dup_path)
        print(count, end=',')
#         data.append(pd.read_csv(dup_path, sep = '\t', engine = 'python', decimal = ',', skiprows = 40, usecols = poss_columns)[2: ].apply(pd.to_numeric))
        data.append(pd.read_csv(dup_path, sep = '\t', engine = 'python', decimal = ',', skiprows = 40)[2: ].apply(pd.to_numeric))
        count += 1
        dup_path = path
    print("No. of files processed are: ", count)
    return data
```

```python
file_data = read_textfiles('/home/loki/MAS/R_D/Moses/Full_data/sk_folded/0031678-8801103(001)/')
```

```python
one_file = file_data[0]
print(len(list(one_file)))
```

```python
columns = list(one_file)
plt.figure(figsize=(30, 30))
for each_column in range(0, 32):
    ylbl = str(columns[each_column])
    plt.subplot(8, 4, each_column+1)
    y = one_file[columns[each_column]]
    plt.scatter(range(0, len(y)), y)
    plt.ylabel(ylbl)
    plt.xlabel("Number of Samples")
    plt.grid()
plt.savefig("Features_full.png", dpi=600)
plt.show()
```

```python
plt.figure(figsize=(30, 30))
for each_column in range(32, 63):
    ylbl = str(columns[each_column])
    plt.subplot(8, 4, each_column-32+1)
    y = one_file[columns[each_column]]
    plt.scatter(range(0, len(y)), y)
    plt.ylabel(ylbl)
    plt.xlabel("Number of Samples")
    plt.grid()
plt.savefig("Features_full_2.png", dpi=600)
plt.show()
```

### Removed columns

```python
removable_columns = ['Lngsbesch', 'GPS_Status', 'Status_DIS', 'Status_TRK', 'GPS_FMS_St','Breitengrd', 'Lngengrad', 
                     'Breitengrd', 'Hhe', 'Fahrg_NULL', 'StWhl_AnglSpd', 'Odo', 'AirTemp_Outsd_Disp', 'AirTemp_Outsd',
                     'BrkPdl_Stat', 'Brk_Stat', 'FullBrk_Actv', 'BrkIntrvntn_Actv_ESP', 'ESP_CtrlLmp_Info',
                     'AccelPdlPosn_Raw', 'AccelPdlPosn', 'KickDnSw_Psd',  'TankLvl', 'TankLvl_Disp_V2', 'SupBat_Volt', 
                     'SupBat_Curr', 'StW_Trq_EPS', 'EPS_FtWhlAngl', 'EPS_FtWhlAngl_Offset', 'VehAccel_X_Offset',
                      'VehYawRateOffset_V2', 'BrkTrq_V2', 'VehAccel_Y_Offset']
## Remove the columns whose values don't change
for each_file in range(len(file_data)):
    file_data[each_file] = file_data[each_file].drop(columns=removable_columns)
print(list(file_data[0]))
print(len(list(file_data[0])))
```

```python
one_file = file_data[0]
print(len(list(one_file)))
```

```python
columns = list(one_file)
count = 0
while(count<len(columns)):
    plt.figure(figsize=(12, 12))
    x = count
    for each_column in range(count, count+4):
        if each_column < len(columns):
            ylbl = str(columns[each_column])
            plt.subplot(2, 2, each_column+1-x)
            y = one_file[columns[each_column]]
            plt.plot(range(0, len(y)), y, label = columns[each_column])
            plt.ylabel(ylbl, fontsize = 12)
            plt.xlabel("Number of Samples", fontsize = 12)
            plt.grid()
            plt.gca().legend(fontsize = 12)
        else:
            break
    Title = "Features_changed_"+str(count)+".png"
    plt.savefig(Title)
    plt.show()
    count += 4
```

```python
plt.figure(figsize=(30, 30))
for each_column in range(16, len(columns)):
    ylbl = str(columns[each_column])
    plt.subplot(4, 4, each_column-16+1)
    y = one_file[columns[each_column]]
    plt.plot(range(0, len(y)), y)
    plt.ylabel(ylbl)
    plt.xlabel("Number of Samples")
    plt.grid()
plt.savefig("Features_changed_2.png", dpi=600)
plt.show()
```
