```python
header_data = pd.read_csv('/home/loki/MAS/R_D/Moses/git/R-D-moses/CSV Files/concat.csv', header=None, engine='python', skiprows = [28])
header_data = header_data.drop(columns= [0]).fillna(-1.1111)
header_data.columns = range(header_data.shape[1])
```

```python
name_data = pd.read_csv('/home/loki/MAS/R_D/Moses/git/R-D-moses/CSV Files//concat_name.txt', header=None, engine='python')
```

```python
header_data
```

```python
header_data = header_data.transpose()
header = header_data.to_numpy().tolist()
name_data = name_data.transpose()
name = name_data.to_numpy().tolist()
```

```python
proc_header = []
unique_header = []
for i in header:
    proc_header.append(",".join(str(j) for j in i))
zipped = list(zip(proc_header, name))
for i in proc_header:
    if i not in unique_header:
        unique_header.append(i)
```

```python
print(zipped[0][1])
print(unique_header[0])
```

```python
header_list =[]
for i in unique_header:
    temp = []
    for j in zipped:
        if i == j[0]:
            temp.append(j[1][0])
    header_list.append("/".join(temp))
```

```python
num_list = []
for i in unique_header:
    num_list.append(proc_header.count(i))
```

```python
six = []
eight = []
four = []
sixteen = []
fourteen = []
tweleve = []
irregular = []
for i in range(len(num_list)):
    if num_list[i] == 6:
        six.append(i)
    elif num_list[i] == 8:
        eight.append(i)
#     elif num_list[i] == 4:
#         four.append(i)
    elif num_list[i] == 16:
        sixteen.append(i)
    elif num_list[i] == 14:
        fourteen.append(i)
    elif num_list[i] == 12:
        tweleve.append(i)
    else:
        irregular.append(i)
    
```

```python
print(len(six))
print(len(eight))
print(len(four))
print(len(sixteen))
print(len(fourteen))
print(len(tweleve))
print(len(irregular))
```
